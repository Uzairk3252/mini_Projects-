```python
import speech_recognition as sr
import os
import subprocess
import pyautogui
import time

# Initialize recognizer
r = sr.Recognizer()

# Function to create a new text file and open it in Notepad
def open_new_file():
    # Create a new file path for a temporary text file
    temp_file = "temp_file.txt"
    with open(temp_file, "w") as file:
        file.write("")  # Create an empty file
    
    # Open the new file in Notepad
    subprocess.Popen(["notepad.exe", temp_file])

    # Give Notepad some time to open
    time.sleep(2)
    
    return temp_file

# Function to listen for the file name after "save" command
def listen_for_filename():
    print("Listening for file name...")
    with sr.Microphone() as source:
        r.adjust_for_ambient_noise(source, duration=0.5)
        audio = r.listen(source)

        try:
            # Recognize speech for file name
            file_name = r.recognize_google(audio).lower()
            print(f"File name recognized: {file_name}")
            return file_name
        except sr.UnknownValueError:
            print("Could not understand the file name, using default 'untitled'")
            return "untitled"
        except sr.RequestError as e:
            print(f"Error requesting file name from Google Speech Recognition service; {e}")
            return "untitled"

# Function to listen and write to Notepad
def listen_and_write():
    open_new_file()  # Create and open the temporary file
    
    while True:
        try:
            with sr.Microphone() as source:
                print("Listening...")
                # Adjusting for background noise quickly (set lower duration)
                r.adjust_for_ambient_noise(source, duration=0.5)
                
                # Listen with a phrase time limit for quicker processing
                audio = r.listen(source, timeout=5, phrase_time_limit=5)

                # Recognize speech using Google's speech recognition
                text = r.recognize_google(audio).lower()  # Convert to lowercase for command checking
                print("You said: " + text)

                # Check for commands
                if "save" in text:
                    # Ask for the file name via voice recognition
                    file_name = listen_for_filename() + ".txt"
                    # Simulate saving the file with the provided name
                    pyautogui.hotkey('ctrl', 's')
                    time.sleep(1)  # Small delay before typing the file name
                    
                    # Type the file name into the Save As dialog and press Enter
                    pyautogui.write(file_name, interval=0.05)
                    pyautogui.press('enter')
                    print(f"Saving the file as '{file_name}' and exiting...")
                    break
                elif "next line" in text:
                    # Move to the next line in Notepad
                    pyautogui.press('enter')
                    print("Moving to next line...")
                elif "next paragraph" in text:
                    # Add an empty line to start a new paragraph
                    pyautogui.press('enter')
                    pyautogui.press('enter')
                    print("Starting a new paragraph...")
                else:
                    # Set a faster typing speed for pyautogui (reduce delay between keypresses)
                    pyautogui.write(text, interval=0.01)  # Typing very fast

        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand audio")
        except sr.RequestError as e:
            print(f"Could not request results from Google Speech Recognition service; {e}")
        except sr.WaitTimeoutError:
            print("Listening timed out, try again.")

if __name__ == "__main__":
    listen_and_write()
```

```python
import speech_recognition as sr
import os
import subprocess
import pyautogui
import time

# Initialize recognizer
r = sr.Recognizer()

# Function to create a new text file and open it in Notepad
def open_new_file():
    # Create a new temporary file path
    temp_file = "temp_file.txt"
    with open(temp_file, "w") as file:
        file.write("")  # Create an empty file
    
    # Open the new file in Notepad
    try:
        subprocess.Popen(["notepad.exe", temp_file])
        print("Notepad opened successfully.")
    except Exception as e:
        print(f"Failed to open Notepad: {e}")

    # Give Notepad some time to open
    time.sleep(2)
    
    return temp_file

# Function to listen and write to Notepad
def listen_and_write():
    open_new_file()  # Create and open the temporary file
    
    while True:
        try:
            with sr.Microphone() as source:
                print("Listening...")
                # Adjusting for background noise quickly (set lower duration)
                r.adjust_for_ambient_noise(source, duration=0.5)
                
                # Listen with a phrase time limit for quicker processing
                audio = r.listen(source, timeout=5, phrase_time_limit=5)

                # Recognize speech using Google's speech recognition
                text = r.recognize_google(audio).lower()  # Convert to lowercase for command checking
                print("You said: " + text)

                # Check for commands
                if "exit" in text:
                    # Exit command to close Notepad
                    pyautogui.hotkey('alt', 'f4')
                    print("Exiting Notepad and program...")
                    break
                elif "stop" in text:
                    # Stop the entire program
                    print("Stopping the program completely...")
                    return  # Exit the function and stop the entire program
                elif "next line" in text:
                    # Move to the next line in Notepad
                    pyautogui.press('enter')
                    print("Moving to next line...")
                elif "next paragraph" in text:
                    # Add an empty line to start a new paragraph
                    pyautogui.press('enter')
                    pyautogui.press('enter')
                    print("Starting a new paragraph...")
                elif "space" in text:
                    # Insert a space in Notepad
                    pyautogui.press('space')
                    print("Inserting a space...")
                elif "backspace" in text:
                    # Delete the last word
                    pyautogui.hotkey('ctrl', 'backspace')
                    print("Deleting the last word...")
                else:
                    # Set a faster typing speed for pyautogui (reduce delay between keypresses)
                    pyautogui.write(text, interval=0.01)  # Typing very fast

        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand audio")
        except sr.RequestError as e:
            print(f"Could not request results from Google Speech Recognition service; {e}")
        except sr.WaitTimeoutError:
            print("Listening timed out, try again.")

if __name__ == "__main__":
    listen_and_write()
```
